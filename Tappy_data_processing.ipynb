# Generate users dataframes (demographic stat + Tappy data)

```python
user_list_pd = pd.read_csv("Tappy_users_id.txt", header=None)
user_list = user_list_pd[0].tolist()
```

```python
col_names = ["User_id",
             "BirthYear",
             "Gender",
             "Parkinsons",
             "Tremors",
             "DiagnosisYear",
             "Sided",
             "UPDRS",
             "Impact",
             "Levadopav",
             "DA",
             "MAOB",
             "Other"]

users_stat = pd.DataFrame(columns = col_names)
for i in user_list:
    
    curr_user_id = re.split(r'_|.txt', i)[1]
    
    curr_user = []
    curr_user.append(curr_user_id)
    
    for line in open('tappy-keystroke-data-1.0.0/Archived users/'+i):
        
        curr_user.append(line.strip().split(':')[1])
        
    users_stat = users_stat.append(pd.DataFrame([curr_user], columns=col_names), ignore_index=True)
```

```python
# UserKey: 10 character code for that user
# Date: YYMMDD
# Timestamp: HH:MM:SS.SSS
# Hand: L or R key pressed
# Hold time: Time between press and release for current key mmmm.m milliseconds
# Direction: Previous to current LL, LR, RL, RR (and S for a space key)
# Latency time: Time between pressing the previous key and pressing current key. Milliseconds
# Flight time: Time between release of previous key and press of current key. Milliseconds

col_names_data = ["User_id",
             "Date",
            "Timestamp",
            "Hand",
            "Hold_Time",
            "Direction",
            "Latency_Time",
            "Flight_Time"]
```

```python
user_data = pd.read_csv("merged_data.txt", names = col_names_data, sep = "\t", index_col=False)
```
